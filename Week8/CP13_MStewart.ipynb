### Exercise \#3

```python
# Scale as given in notes
sc = preprocessing.StandardScaler()
input_data = sc.fit_transform(trainingDFcut.iloc[:, :-1])
target = trainingDFcut.iloc[:, -1]
# set random seed
Answer_to_all_questions = 42
# train - test split of dataset
train_data, test_data, train_target, test_target = model_selection.train_test_split(\
    input_data, target, test_size = 0.3, random_state = Answer_to_all_questions)
print(train_data.shape, train_target.shape, test_data.shape, test_target.shape)
```

### Exercise \#4

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# Model architecture from notes
def my_model(num_inputs, num_nodes, extra_depth):
    # create model
    model = Sequential()
    model.add(Dense(num_nodes, input_dim = num_inputs, kernel_initializer = 'normal', \
        activation = 'relu'))
    model.add(Dropout(0.2))
    for i in range(extra_depth):
        # code up the extra layers here
        model.add(Dense(num_nodes, input_dim = num_nodes, kernel_initializer = 'normal', \
            activation = 'relu'))
    # Output is one probability of event being a signal event
    num_outputs = 1
    model.add(Dense(num_outputs, activation = 'sigmoid'))
    # Compile model
    model.compile(loss = 'binary_crossentropy' , optimizer = 'adam' , metrics =['accuracy'])
    return model
```

```python
batchSize = 500
N_epochs = 20
```

```python
model1 = my_model(train_data.shape[1], 20, 1)
```

```python
history1 = model1.fit(train_data, train_target, batch_size = batchSize, epochs = N_epochs, \
    verbose =1, validation_data =(test_data, test_target))
```

```python
nn_utils.lossplot(history1.history['loss'], scale='log')
```

```python
nn_utils.accplot(history1.history['val_accuracy'], scale='linear')
```

Successful. Loss decays nicely until plateau and accuracy is high. 

### Exercise \#5

```python
# For data with 'reco_zv_mass' back in
sc = preprocessing.StandardScaler()
input_data2 = sc.fit_transform(trainingDF.iloc[:, :-1])
target2 = trainingDF.iloc[:, -1]
# set random seed
Answer_to_all_questions = 42
# train - test split of dataset
train_data2, test_data2, train_target2, test_target2 = model_selection.train_test_split(\
    input_data2, target2, test_size = 0.3, random_state = Answer_to_all_questions)
print(train_data2.shape, train_target2.shape, test_data2.shape, test_target2.shape)
```

```python
def new_model(num_inputs, num_nodes, extra_depth):
    # create model
    model = Sequential()
    model.add(Dense(num_nodes, input_dim = num_inputs, kernel_initializer = 'normal', \
        activation = 'relu'))
    model.add(Dropout(0.2))
    for i in range(extra_depth):
        # code up the extra layers here
        model.add(Dense(num_nodes, input_dim = num_nodes, kernel_initializer = 'normal', \
            activation = 'relu')) 
    num_outputs = 1
    model.add(Dense(num_outputs, activation = 'sigmoid'))
    # Compile model
    model.compile(loss = 'binary_crossentropy' , optimizer = 'adam' , metrics =['accuracy'])
    return model
```

```python
model2 = new_model(train_data2.shape[1], 40, 4)
```

```python
batchSize = 500
# With early stopping, we can set an arbitrarily long time here
N_epochs = 100
```

```python
callbacks_ = [
# if we don ’t have an increase of the accuracy for 10 epochs , terminate training .
EarlyStopping(verbose = True, patience =10, monitor = 'val_accuracy') ,
# Always make sure that we ’ re saving the model weights with the best accuracy .
ModelCheckpoint('model.h5', monitor = 'val_accuracy', verbose =0, save_best_only = True, mode = 'max')]
history2 = model2.fit(train_data2, train_target2, batch_size = batchSize, epochs = N_epochs, \
    verbose =1, validation_data =(test_data2, test_target2), callbacks = callbacks_ )
```

```python
nn_utils.lossplot(history2.history['loss'], scale='log')
```

```python
nn_utils.accplot(history2.history['val_accuracy'], scale='linear')
```

Accuracy reaches 95\% ... to 2 sig. fig :)

### Exercise \#6

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
# Get probabilities of signal event for test set
test_pred = model2.predict(test_data2)
# Round >= 0.5 up to 1 and < 0.5 down to 0
predicted = np.round(test_pred)
```

```python
# Confusion matrix
cm = confusion_matrix(test_target2, predicted, normalize="true")
fig, ax = plt.subplots(figsize=(6, 6))
sns.heatmap(cm, cmap=plt.cm.Blues,annot=True,fmt=".2f")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")
ax.set_yticklabels(['backgound', 'signal'])
ax.set_xticklabels(['backgound', 'signal'])
plt.title('Confusion matrix ')
plt.show()
```

About 95\% accurate for both signal and background events.

```python
# ROC curve
fpr, tpr, thresholds = roc_curve(test_target2, test_pred)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
lw = 2
plt.plot(
    fpr,
    tpr,
    color="darkorange",
    lw=lw,
    label="ROC curve (area = %0.2f)" % roc_auc,
)
plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver operating characteristic")
plt.legend(loc="lower right")
plt.show()
```

Curve area near 1 and graph in upper left mostly so looks good.

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(thresholds[1:], tpr[1:], label='True positive rate')
ax.plot(thresholds[1:], fpr[1:], label='False positive rate')
ax.set_xlabel('threshold')
ax.set_ylabel('rate value')
ax.legend()
plt.show()
```

At threshold of 0.2 and above for values to be considered signal, fpr has dropped significantly. After threshold of 0.8, tpr  drops significantly. So any threshold between 0.2 and 0.8 works well for this model.

At threshold of 0, everything is considered signal so fpr and tpr are both 1 while at threshold of exactly 1, nothing is considered signal so fpr and tpr are both 0 as expected.
